```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = pd.read_csv(r"C:\Users\Kumar\Downloads\customer_purchase_behavior.csv")

df.head()
```

```python
df.info()
```

```python
# Histogram of Spending Score
plt.figure(figsize=(8,5))

sns.histplot(df['SpendingScore'], bins=10)

plt.title("Histogram of Spending Score")
plt.xlabel("Spending Score")
plt.ylabel("Frequency")

plt.show()
```

```python
# Histogram of AnnualIncome
plt.figure(figsize=(8,5))

sns.histplot(df['AnnualIncome'], bins=10)

plt.title("Histogram of Annual Income")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")

plt.show()
```

```python
# Boxplot of SpendingScore by CustomerType
plt.figure(figsize=(8,5))

sns.boxplot(
    x='CustomerType',
    y='SpendingScore',
    data=df
)

plt.title("Spending Score by Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Spending Score")

plt.show()
```

```python
# Scatter Plot: Age vs SpendingScore
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='AnnualIncome',
    y='SpendingScore',
    hue='CustomerType',
    data=df
)

plt.title("Annual Income vs Spending Score")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")

plt.show()
```

```python
# Choose One → Correlation Heatmap
plt.figure(figsize=(6,4))

sns.heatmap(
    df[['Age', 'AnnualIncome', 'SpendingScore']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```
